```python
import pandas as pd
```

```python
train_file = '../data/en_ewt-up-train.conllu'
test_file = '../data/en_ewt-up-test.conllu'
```

```python
with open(train_file, 'r', encoding='utf-8') as f:
    train_data = f.read()
```

```python
data = []
for doc_i, doc in enumerate(train_data.split('\n\n')[:2]):
    doc = doc.split('\n')
    sentences = ''
    for line in doc:
        if line and line[0] != '#':
            line = line.split('\t')
            
#             if len(line) > 11:
#                 for i in range(11, len(line)):
#                     print('create sent for each', line[i])
#                 print('\n', doc)
            line.insert(0, str(doc_i))
            # line[0] = f'{doc_i}_{line[0]}'
            sentences += '\t'.join(line) + '\n'
    data.append(sentences)
```

```python
train_df = pd.DataFrame([x.split('\t') for sent in data for x in sent.split('\n') if x])
```

```python
train_df.shape
```

```python
train_df = train_df.rename(columns={0:'sent_id', 1: 'token_id', 2: 'token', 3: 'lemma', 4: 'POS', 5: 'uni_POS',
                                   6: 'morph_type', 7: 'distance_head', 8: 'dep_label', 9: 'dep_rel', 10: 'space', 11: 'probbank'})
```

```python
def split_sentences(row):
    # print('\n', row['sent_id'].values, '\n')
    sent_id = row['sent_id'].values[0]
    # print(sent_id[0])
    sent = train_df[train_df['sent_id'] == sent_id]
    all_target_columns = sent[list(range(12, train_df.shape[1]))]
    target_columns = all_target_columns[all_target_columns == 'V'].dropna(axis=1, how='all')
    # print(target_columns.columns)
    
    for i in target_columns.columns:
        train_df.append(pd.DataFrame(train_df.iloc[target_columns.index], columns=train_df.columns+[i]))
    
```

```python
train_df.groupby('sent_id').apply(split_sentences)
```
