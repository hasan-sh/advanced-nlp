```python
import pandas as pd
```

```python
train_file = '../data/en_ewt-up-train.conllu'
test_file = '../data/en_ewt-up-test.conllu'
```

```python
with open(train_file, 'r', encoding='utf-8') as f:
    train_data = f.read()
```

```python
data = []
for doc_i, doc in enumerate(train_data.split('\n\n')[:100]):
    doc = doc.split('\n')
    sentences = ''
    for line in doc:
        if line and line[0] != '#':
            line = line.split('\t')
            
#             if len(line) > 11:
#                 for i in range(11, len(line)):
#                     print('create sent for each', line[i])
#                 print('\n', doc)
            line[0] = f'{doc_i}_{line[0]}'
            sentences += '\t'.join(line) + '\n'
    data.append(sentences)
```

```python
train_df = pd.DataFrame([x.split('\t') for sent in data for x in sent.split('\n') if x])
```
